```python
from scipy.stats import chi2_contingency
chitable=pd.crosstab(datasheet.Attrition,datasheet.TotalWorkingYears)
chitable
```

```python
stats,p,dof,expected=chi2_contingency(chitable)
```

```python
print(stats,p)
```

H0: Attrition Rate and TotalWorkingYears are not dependent on each other
Result: Reject H0
