# 2.Create a 5x5 matrix with values 1,2,3,4 just below the diagonal

```python
import numpy as np
data=np.zeros((5,5))
print(data)
i=1
while i>0 and i<5:
    data[i][i-1]=i
    i=i+1
print("final matrix:\n",data)
```

# 3.Create a 8x8 matrix and fill it with a checkerboard pattern

```python
i=0
j=0
import numpy as np
data=np.zeros((8,8))
for i in range(8):
    for j in range(8):
        if (i+j)%2!=0:
            data[i][j]=1
print(data)
            
```

# 4. Normalize a 5x5 random matrix

```python
import numpy as np
data= np.random.rand(5,5)
print("Before normalization:")
print(data)
max = data.max()
min = data.min()
data = (data - min)/(max - min)
print("After normalization:")
print(data)
```

# 5.  How to find common values between two arrays?

```python
import numpy as np
data=np.array([[1,2,3],[4,5,6]])
data1=np.array([[1,8,9],[2,11,4]])
data2=np.intersect1d(data,data1)
print("1st array:",data)
print("2nd array:",data1)
print("common elements:",data2)
```

# 6.How to get the dates of yesterday, today and tomorrow?

```python
import numpy as np
today=np.datetime64('today','D')
yesterday=np.datetime64('today','D')-np.timedelta64(1,'D')
tomorrow=np.datetime64('today','D')+np.timedelta64(1,'D')
print("yesterday:",yesterday)
print("today:",today)
print("tomorrow:",tomorrow)
```

# 8.Create random vector of size 10 and replace the maximum value by 0 

```python
import numpy as np
data=np.random.rand(10)
print(data)
data[np.where(data == np.amax(data))]=0
print(data)
```

# 12.How to get the diagonal of a dot product?

```python
import numpy as np
a=np.arange(25).reshape(5,5)
print("array1:\n",a)
b=np.arange(25).reshape(5,5)
print("array2:\n",b)
c=a.dot(b)
print("dot product:\n",c)
d=c.diagonal()
print("diagonal elemts of dot product:\n",d)
```

# 15.How to create a record array from a regular array?

```python
import numpy as np
a1=np.array([1,2,3,4,5])
a2=np.array(["sector1","sector2","sectror3-dwarka","sectror3-surendranagar","kashmiri gate"])
a3=np.array([12.20,15,20,40,70])
result= np.core.records.fromarrays([a1, a2, a3],names='a,b,c')
print(result[0])
print(result[1])
print(result[2])
print(result[3])
print(result[4])
print(result)
```
